```python
'''
A classe de funcionários contém os dados que serão compartilhados por suas classes de crianças.
A classe do funcionário contém um construtor para rubricar seus atributos.

Os dados utilizados são:
  -nome completo,
  -número de identificação e número de telefone.
'''


# Superclass
class Funcionarios():

  def __init__(self, nome_completo, numero_de_identificacao,numero_de_telefone):
    self.nome_completo = nome_completo
    self.numero_de_identificacao = numero_de_identificacao
    self.numero_de_telefone = numero_de_telefone

  def nome_completo(self):
    return {self.nome_completo}

  def numero_de_identificacao(self):
    return {self.numero_de_identificacao}

  def numero_de_telefone(self):
    return {self.numero_de_telefone}
```

```python
'''
Tipo de funcionário de prazo definido.

Esta classe herda da classe dos funcionários.
Os novos atributos são:
• Número de postos
• Salário básico
• Duração do contrato em meses

Além disso, tem um método que calcula o salário total.
O empregado recebe um aumento de 2% sobre seu salário base.
'''

#Subclasse
class FuncionariosComPrazo(Funcionarios):
  #Subclass
  def __init__(self, nome_completo, numero_de_identificacao,numero_de_telefone,numero_de_postos,salario_basico,duracao_do_contrato,salario_liquido):
    # Atributo da classe FuncionarioDePrazo
    self.numero_de_postos = numero_de_postos
    self.salario_basico = salario_basico
    self.duracao_do_contrato = duracao_do_contrato
    self.salario_liquido = salario_liquido
    self.calcula_o_salario_total = self.salario_basico + (salario_basico * 0.02)

    # Atributo da classe Funcionarios
    Funcionarios.__init__(self,nome_completo,numero_de_identificacao,numero_de_telefone)

    def numero_de_postos(self):
      return self.numero_de_postos

    def salario_basico(self):
      return self.salario_basico

    def duracao_do_contrato(self):
      return self.duracao_do_contrato

    def calcula_o_salario_total(self):
      return self.salario_basico + (self.salario_basico * 0.02)
```
