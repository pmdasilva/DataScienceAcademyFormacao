```python
#Crie a classe Smartphone
class Smartphone():
    # Crie a classe Smartphone com 2 atributos, tamanho e interface
    def __init__(self, tamanho, interface):
        self.tamanho = tamanho
        self.interface = interface
        print('construtor smartphone criado !')

    smart = Smartphone('3629px', 'android')
    print(smart.tamanho)
    print(smart.interface)
```

```python
#crie a subclasse MP3Player
class Mp3Player(Smartphone):
    #com os atributos capacidade
    def __init__(self, tamanho, interface, capacidade):
        super().__init__(tamanho, interface)
        self.capacidade = capacidade
        print('construtor MP3Player criado !')


    mp3 = Mp3Player('3629px', 'android', '128gb')
    print(mp3.capacidade)
```
