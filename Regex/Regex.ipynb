# Lab 3 - Trabalhando com Expressões Regulares em Python com ChatGPT

Expressões regulares são uma sequencia de caracteres que definem uma busca padrão em strings. Em python, as Expressões regulares são suportadas pelo pacaote (re). Ele fornece uma serie de funções para pesquisar e substituir padrões em strings. Algumas das tarefas mais comuns que podem ser realizadas com expressões regulares incluem verificar se uma string corresponde a um determinado padrão, extrair informações e uma string com base em um padrão especifico e substituir trechos de uma string com base em um padrão.

Por exemplo, você pode usar uma expressão regular para verificar se uma string representa um endereço de e-mail valido, ou para encontrar todas as ocorrencias de um padrão em uma string.

### Expressões Regulares

Expressoes regulares são padrões para combinar ou encontrar ocorrencias de sequencias de string, Em Python, expressões regulares são geralemnte usadas para manipular strings e realizar tarefas como validação de netrada de dados, extração de informações de string e substituição de texto.

```python
import re # Regular expression
```

```python
texto = 'Meu e-mail é exemplo@gmail.com e você pode me contar em outro_exemplo@gmail.com'
```

```python
# Expressões regulares para contar quantas vezes o caracter arroba aparece no texto.
resultado = len(re.findall('@',texto))
```

```python
print(f'O caractere "@" apareceu, {resultado} vezes no texto')
```

```python
# Expressão regular para extrair a palavra que aparece após a palavra 'voce' em um texto
resultado = re.findall(r'você (\w+)', texto)
```

```python
print("A palavra após 'você' é: ", resultado[0])
```

Nota: O r antes da string que representa a expressão regular em Python é usado para indicar que a string é uma string literal raw. Isso significa que as barras invertidas () são interpretadas como caracteres de escape, mas são incluidas na expresão regular como parte do padrão.

```python
# expressão regular para extrair endereços de e-mail de uma string
emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', texto)
```

```python
print(emails)
```

```python
text = 'O aluno estava incrivelmente perdido, mas encontrou a DSA e rapidamente começou a aprender'
```

```python
# Extraindo os adverbios da frase
for m in re.findall(r'\w=mente\b',text):
  print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

##### 1 - Crie um REGEX para contar quantas vezes o caracter 'a' aparece em todo o texto da musica.

#### Exemplo 1 = usando r'a' que corresponde a letra 'a' dentro da letra da musica.

```python
# 1 - Crie um REGEX para contar quantas vezes o caracter 'a' aparece em todo o texto da musica.
def conte_as_letras_a():
  try:
    conta_as_letras_a = len(re.findall(r'a', letra_da_musica))
    return f'A letra "a" aparece: {conta_as_letras_a} vezes na musica.'
  except:
    return f'As letras não foram encontradas.'

conte_as_letras_a()
```

```python
# 2 - Crie um REGEX em python para contar quantas vezes as palavra these aparece na musica.
def conte_as_palavras_these():
  try:
    conte_a_palavra = len(re.findall(r'\bthese\b', letra_da_musica))
    return f'A palavra "these" aparece na letra: {conte_a_palavra} vezes. '
  except:
    return f'As palavras não foram encontradas.'

conte_as_palavras_these()
# \b => Sequencia de delimitação de palavra, que garante que o regex corresponda apenas á palavra completa 'these' e não partes de outras palavras.
```

```python
# 3 - Crie um REGEX em python para extrair as palavras seguidas por exclamação.
def extrair_pontos_interrogacao():
  try:
    conte_os_pontos_de_interrogacao = len(re.findall(r'\w+\?', letra_da_musica))
    return f'Na letra existe o numero de: {conte_os_pontos_de_interrogacao} vezes.'
  except:
    return f'Os carecteres especiais não foram achados.'

extrair_pontos_interrogacao()

# \? => Corresponde ao caracter de ? de forma literal.
```

```python
# 4 - Crie um REGEX que extrai qualquer palavra cujo antecessor seja a palavra 'these' e o sucessor seja 'lies'.
def antecessor_e_sucessor():
  try:
    pegue_a_palavra = re.findall(r'\ball\s(\w+)\slies\b', letra_da_musica)
    print(f'a palavra que esta entre all e lies é: {pegue_a_palavra[0]}')
  except:
    print('As palavras não foram encontradas.')

antecessor_e_sucessor()
```
