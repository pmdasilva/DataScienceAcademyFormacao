#### While e For Juntos

- Vamos encontrar número primos em uma coleção usando loop While e For Juntos.

Um número primo é um numero natural maior do que 1 é divisivél apenas por ele mesmo. Isso significa que não há outro numero inteiro que possa dividir o número primo sem deixar restos. por exemplo, o numero 2 é primo, pois é divisivél por 1 e 2. O numero 4 não é primo, pois é divisivel por 2.

-- Segue o pseudocódigo

1. Inicialize uma lista vazia para armazenar os numeros primos
2. Para cada numero N entre 2 e 30:
  Inicialize uma variavél eh_primo com verdadeiro
  Para cada número i entre entre 2 e N/2:
   Se N é divisivél por i, então:
   Altere a variavél eh_primo para falso
   Pare de verificar os outros números
   Se a variavél eh_primo ainda é verdadeira, adicione N á lista de números primos
  Imprime a lista de numeros primos




```python
lista_de_numeros_primos = []

for num in range(2,31):
  eh_primo = True

  #Verificar se o numero é primo
  i = 2
  while i <= num // 2:
    if num % i == 0:
      eh_primo = False
      break
    i += 1

  #Adicionando um numero primo a array:
  if eh_primo:
    lista_de_numeros_primos.append(num)

  #Mostrando a lista com numero primos
  print(lista_de_numeros_primos)
```

```python
# Faça um programa que leia e valide as seguintes informações:

def main():

  # Nome: maior que 3 caracteres;
  def valida_nome():
    nome = input('Digite seu nome: ')
    while True:
      if len(nome) < 3:
        print('Nome invalido digite novamente !')
        nome = input('Digite seu nome: ')
      else:
        print(f'O nome informado é {nome}')
        break

  # Idade: entre 0 e 150;
  def valida_idade():
    while True:
      idade = int(input('Digite sua idade: '))
      if idade >= 0 and idade <= 150:
        print(f'A idade é {idade}')
        break
      else:
        print('Idade invalida digite novamente sua idade:')



  # Salário: maior que zero;
  def valida_salario():
    while True:
      salario = float(input('Digite seu salario: '))
      if salario > 0:
        print(f'O salario é de: {salario}')
        break
      else:
        print('Salario invalido digite novamente seu salario')

  # Sexo: 'f' ou 'm';
  def valida_sexo():
    while True:
      sexo = str(input('Digite o seu sexo "m" ou "f": '))
      if sexo == 'm' or sexo == 'f':
        print(f'o sexo do usuario é: ({sexo})')
        break
      else:
        print('Sexo invalido digite novamente seu sexo')

  # Estado Civil: 's', 'c', 'v', 'd';

  def valida_estao_civil():
    while True:
      estado_civil = str(input('Digite seu estado civil "s" "c" "v" "d": '))

      if estado_civil == 's' or estado_civil == 'c' or estado_civil == 'v' or estado_civil == 'd':
        print(f'O estado civel é: {estado_civil}')
        break
      else:
        print('Estado civil invalido digite novamente seu estado civil')

  # Use a função len(string) para saber o tamanho de um texto (número de caracteres).]
  def contador_de_caractere():
    while True:
      texto = input('Digite um texto: ')
      if texto == '':
        print('Texto invalido digite novamente seu texto')
      else:
        print(f'O texto possui {len(texto)} caracteres')
        break

  # Chamando as funções:
  def call_funcao():
    valida_nome()
    print(f'*' * 50)
    valida_idade()
    print(f'*' * 50)
    valida_salario()
    print(f'*' * 50)
    valida_sexo()
    print(f'*' * 50)
    valida_estao_civil()
    print(f'*' * 50)
    contador_de_caractere()
    print(f'*' * 50)

  call_funcao()


main()



















```
