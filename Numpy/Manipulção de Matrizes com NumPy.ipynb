# Analise estatisticas Básica com NumPy - Desvio Padrão

O desvio padrão é uma medida estatistica de dispersão que inidca o quanto os valores de um conjunto de dados se afastam da media. Ele é calculado como a raiz quadrada da variancia, que é a média dos quadrados das diferencias entre cada valor e a média.

O desvio padrão é um medida util porque avaliar a variedade dos dados em tomo da média. Se os valores estivesem proximo da média, o desvio padrão será baixo, indicando que os dados tem pouca variabilidade. Por outro lado, se os valores estiverem muito diferente da média, o desvio padrão será alto, indicando que os dados tem alta variabilidade.

O desvio padrão é amplamente utilizado em analise e ciencia de dados, para avaliar a consistencia dos dados conjuntos de dados diferentes. É importante notar que o desvio padrão pode ser influenciado por valores extremos (outlers) e pode ser afetado por diferentes distribuições de dados.

```python
# Analise estatisticas Básica com NumPy - Desvio Padrão
np.std(arr41)
```

- Quanto menor o valor ,mais proximo os valor de pradronização de valores. E se os valores for muito grande, nos informa que os valores estão muito diferentes do normal.

# Analise Estatistica Básica com NumPy - Variancia

- A variancia é uma medida estatistica que qualifica a dispersão dos valores em um conunto de dados em relação á média. Ela é calculada como a média dos quadrados das diferenças entre cada valor e a média.

- A formula para calculo das variaveis é:

var = 1/n*2(x) - x^2

Onde:
- var é variavel
- n é o numero de observações
- z é o somado
- x é o i-éssimo valor no conjunto de dados
- x é o maximo dos valores


- A variancia é uma medida util para avaliar a variabilidade dos dados em torno da média. Se a variancia for baixa, indica que os valores estão proximos da medida e tem pouca variabilidade. Por outro lado, se a variebalidade for alta, isso indica que os valores estão distantes da media e tem alta variedade.

```python
# Variancia
np.var(arr41)
```

#  Quando usar desvio padrao ou variante

Tanto a variante quando o desvio padrão são medidas uteis de dispersão e podem ser usados em conjuto para descrever a distribuição de um conjunto de dados.

A variancia é uma medida quadrática e pode ser útil para calcular outras estatisticas, como o desvio padrão. No entanto, como a variancia é uma medida quadratica, seus valores são geralmente maiores do que os valores dos proprios dados, o que pode dificultar a interpretação. o desvio padrão é a raiz quadrada da variante e fornece uma medida de dispersão que tem a mesma unidade de medida que os proprios dados, facilitando a interpretação e a comparação com outros valores.

Em geral, o desvio padrão é mais comumente usado que a variancia, principalmente porque é mais facil de interpretar. No entanto, e um escolha entre o uso da variancia ou do desvio padrão depende do contexto e do objetivo da análise. Em outros casos, o desvio padrão pode ser uma medida mais apropriada, como quando se pretende avaliar a consistencia dos dados em relação á media e comparar diferentes conjuntos de dados.
