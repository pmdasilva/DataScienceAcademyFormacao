```python
#Superclass
class Veiculo():
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo
    # self.cor = cor
    self.ligado = False
    self.desligado = True
    self.__velocidade_maxima = 300

  def ligar_veiculo(self):
    print(f'Veiculo esta ligando... veiculo ligado !')

  def desligar_veiculo(self):
    print(f'Veiculo desligando... veiculo desligado!')

  def imprimir_informacoes(self):
    print(f'As caracteristica do veiculo é: {self.marca} - {self.modelo} - sua velocidade maxima é {self.__velocidade_maxima}')

#Subclass
class Carro(Veiculo):
  def __init__(self, marca, modelo):
    Veiculo.__init__(self, marca=marca,modelo=modelo)


#Subclass
class Moto(Veiculo):
  def __init__(self, marca, modelo):
    Veiculo.__init__(self, marca=marca,modelo=modelo)

```

```python
moto1 = Moto('kward','4x4')
moto1.ligar_veiculo()
moto1.desligar_veiculo()
moto1.imprimir_informacoes()
```

```python
carro1 = Carro(marca='Mustang',modelo='Ford')
carro1.ligar_veiculo()
carro1.desligar_veiculo()
carro1.imprimir_informacoes()
```
